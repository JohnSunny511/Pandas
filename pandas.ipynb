```python
df = pd.DataFrame({
    'id' : [1,2,3,4],
    'Name' : ['John', 'Joe', 'Jan', 'Jude']
})

df2 = pd.DataFrame({
    'id' : [3, 4, 5, 6],
    'Name' : ['Jacob', 'Jose', 'jack', 'Rose']
})

inres = pd.merge(df, df2, how = 'inner', on = 'id')
inres
```

```python
outres = pd.merge(df, df2, how = 'outer', on = 'id')
outres
```

```python
lefres = pd.merge(df, df2, how='left', on='id')
lefres
```

```python
rigres = pd.merge(df, df2, how='right', on='id')
rigres
```
